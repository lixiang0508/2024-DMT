```python
import pandas as pd

# load the data
data_path = 'data/aggregate_data.csv'
data = pd.read_csv(data_path)

# 显示数据的前几行以及列名
data.head(), data.columns
```

```python
from business_calendar import Calendar, MO, TU, WE, TH, FR
cal = Calendar(workdays=[MO,TU,WE,TH,FR])
```

```python
def is_working_day(date):
    if cal.isbusday(date):
        return True
    else:
        return False
```

```python
data['is_working_day'] = data['time'].map(is_working_day) # add a feature :  if a date is working day
```

```python
import numpy as np
# 计算 33% 和 67% 的百分位数作为分界点
q33 = np.percentile(data['mood'], 33)
q67 = np.percentile(data['mood'], 67)

data['mood_category'] = pd.cut(data['mood'], bins=[data['mood'].min(), q33, q67, data['mood'].max()], labels=[1, 2, 3])
```

```python
from sklearn.model_selection import train_test_split

# 定义自变量和因变量
X = data[['activity',
       'circumplex.arousal', 'circumplex.valence', 'screen', 'appCat.builtin',
       'appCat.communication', 'appCat.main', 'appCat.tools', 'appCat.contact',
       'mood_label', 'is_working_day']]
y = data['mood_category']

# 拆分数据为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train = y_train.dropna()
X_train = X_train.loc[y_train.index]  # 确保 y_train 与 X_train 行对齐
```

```python
#print(X_train.isnull().sum())
print(y_train.isnull().sum())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# 设置决策树回归器
tree = DecisionTreeClassifier(random_state=42)

# 设置要调优的参数
param_grid = {
    'max_depth': [2, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10]
}

# 设置 GridSearchCV
grid_search = GridSearchCV(estimator=tree, param_grid=param_grid, cv=5, scoring='accuracy')

# 运行 GridSearchCV
grid_search.fit(X_train, y_train)

# 输出最佳参数和最佳得分
grid_search.best_params_, grid_search.best_score_
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 预测测试集结果
y_pred = grid_search.predict(X_test)

# 生成混淆矩阵
conf_mat = confusion_matrix(y_test, y_pred)
print("Confusion matrix：")
print(conf_mat)

# 可视化混淆矩阵
fig, ax = plt.subplots(figsize=(8, 8))
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=grid_search.classes_)
disp.plot(cmap='Blues') 
#plot_confusion_matrix(grid_search, X_test, y_test, ax=ax, cmap='Blues')  # cmap 参数可以调整颜色风格
plt.title("Confusion matrix for Decision Tree")
plt.show()
```
