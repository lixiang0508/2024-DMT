## Consider the following two error measures: mean squared error (MSE) and mean absolute error (MAE).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#For the first situation, we need a larger value range
np.random.seed(0)
true_values_1 = np.linspace(0, 10, 100)
predicted_values_1 = true_values_1 + np.random.normal(0, 2, size=(100,))
mse_1 = mean_squared_error(true_values_1, predicted_values_1)
mae_1 = mean_absolute_error(true_values_1, predicted_values_1)

#For the Second situation, we need a smaller value range
np.random.seed(1)
true_values_2 = np.linspace(0, 10, 100)
predicted_values_2 = true_values_2 + np.random.normal(0, 0.1, size=(100,))
mse_2 = mean_squared_error(true_values_2, predicted_values_2)
mae_2 = mean_absolute_error(true_values_2, predicted_values_2)
```

```python
plt.figure(figsize=(12, 6))

# Situation 1 
plt.subplot(1, 2, 1)
plt.plot(true_values_1,true_values_1,'r',label='True')
plt.scatter(true_values_1, predicted_values_1, color='b', label='Predicted')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Situation 1 (in larger range predicted values)')
plt.text(0.5, 9, 'MSE = {:.2f}'.format(mse_1), fontsize=12, color='g')
plt.text(0.5, 8, 'MAE = {:.2f}'.format(mae_1), fontsize=12, color='m')
plt.legend()

# Situation 2 
plt.subplot(1, 2, 2)
plt.plot(true_values_2,true_values_2,'r',label='True')
plt.scatter(true_values_2, predicted_values_2, color='b', label='Predicted')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Situation 2 (in smaller range predicted values)')
plt.text(0.5, 7, 'MSE = {:.2f}'.format(mse_2), fontsize=12, color='g')
plt.text(0.5, 6.5, 'MAE = {:.2f}'.format(mae_2), fontsize=12, color='m')
plt.legend()

plt.tight_layout()
plt.show()

```
